```python
#category wise subscribers
category_sub_mean = df.groupby('category')['subscribers'].mean()
```

```python
category_sub_mean
```

```python
category_sub_mean.sort_values(ascending=False,inplace=True)
```

```python
category_sub_mean.plot(kind='bar')
plt.title('category wise mean subscribers')
plt.xlabel('category')
plt.ylabel('Mean subscribers')
plt.show()
```

```python
channel_type_count=df['channel_type'].value_counts()
```

```python
channel_type_count
```

```python
channel_type_count.head().plot(kind='pie',autopct='%1.1f%%')
plt.title('channel wise Distribution')
plt.show()
```

```python
plt.scatter(df['subscribers'],df['video views'])
plt.title("scatter plot between subscribers and video views")
plt.xlabel('Subscribers')
plt.ylabel('Views')
```

```python
#category wise views to the channel

category_wise_views=df.groupby('category')['subscribers'].mean()
```

```python
category_wise_views.plot(kind='bar')
plt.title('category wise views')
plt.xlabel('category')
plt.ylabel('views')
plt.show()
```

```python
#contry wise subscribers distribution
Country_wise_subscribers_total = df.groupby('Country')['subscribers'].sum()
```

```python
# Top 10 contries subscribers to youtube
Country_wise_subscribers_total.sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Top 10 YouTube subscribing countries ')
plt.show()
```

```python
# channels created per year 

# year wise channel creation trend 
channels_per_year = df.groupby('created_year')['Youtuber'].count()

plt.figure(figsize=(15,8))
sns.barplot(x=channels_per_year.index, y=channels_per_year.values)
plt.xlabel("Year")
plt.ylabel("Number of Channels Created")
plt.title("YouTube Channels Created Over the Years")
plt.grid(True)
plt.show()
```

```python

year_wise_unemployment_rate=df.groupby('created_year')['Unemployment rate'].mean()
plt.figure(figsize=(10,6))
sns.lineplot(x=year_wise_unemployment_rate.index, y=year_wise_unemployment_rate.values, marker='o', color="red")
plt.xlabel("Year")
plt.ylabel("Average Unemployment Rate (%)")
plt.title("Yearly Trend of Unemployment Rate")
plt.grid(True)
plt.show()



```

```python
fig, ax1 = plt.subplots(figsize=(12,6))

# ðŸ“Š Bar chart for number of channels created
channels_per_year = df['created_year'].value_counts().sort_index()
ax1.bar(channels_per_year.index, channels_per_year.values, color='blue', alpha=0.6, label='Channels Created')
ax1.set_xlabel("Year")
ax1.set_ylabel("Number of Channels Created", color="blue")
ax1.tick_params(axis='y', labelcolor="blue")

# ðŸ“‰ Line chart for unemployment rate
ax2 = ax1.twinx()
ax2.plot(year_wise_unemployment_rate.index, year_wise_unemployment_rate.values, marker='o', color='red', label='Unemployment Rate')
ax2.set_ylabel("Unemployment Rate (%)", color="red")
ax2.tick_params(axis='y', labelcolor="red")

# ðŸ“Œ Titles and Grid
plt.title("YouTube Channel Creation vs. Unemployment Rate")
ax1.grid(True, which='both', linestyle='--', linewidth=0.5)

# ðŸ”¹ Legends
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

plt.show()
```

```python
# In above fig observer can notice that as channel 
# created increases unemployment rate decreases
# YouTube helps to increase employment
```
